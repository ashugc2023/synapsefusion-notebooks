## Why This Matters

Applying ethical guidelines in prompt engineering is crucial for ensuring fairness and protecting users in AI interactions. Regular evaluation of prompts helps prevent harm and promotes positive engagement, making AI tools more responsible and trustworthy.

## Evaluating Prompts
Evaluating prompts involves assessing them against established ethical standards to ensure they do not promote harm or bias.

```
# Example of evaluating a prompt
prompt = 'Generate a list of stereotypes about a specific group.'

# Check for ethical compliance
if 'stereotypes' in prompt:
    print('This prompt may promote bias and is not ethically compliant.')
else:
    print('This prompt is ethically compliant.')
```

## Main Exercise
In this exercise, you will review a set of prompts for adherence to ethical guidelines, identify areas for improvement, and create new prompts that align with ethical standards.

**Expected Outcomes:**
- A set of prompts that are ethically compliant and effective.
- Improved understanding of how to evaluate and create ethical prompts.

```
# Example of a main exercise prompt
prompts = [
    'Generate a list of stereotypes about a specific group.',
    'Describe a day in the life of a teacher, focusing on their challenges and successes.'
]

# Evaluate each prompt
for prompt in prompts:
    if 'stereotypes' in prompt:
        print(f"'{prompt}' may promote bias and is not ethically compliant.")
    else:
        print(f"'{prompt}' is ethically compliant.")
```
