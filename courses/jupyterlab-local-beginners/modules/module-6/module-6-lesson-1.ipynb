### Environment Setup

Setting up your development environment involves configuring the tools and libraries you need for your project. This includes creating virtual environments and installing necessary packages.

```
# Example: Creating a virtual environment
import os
os.system('python -m venv my_project_env')
print('Virtual environment created: my_project_env')
```

### Micro-Exercise 2: Set Up Your Environment

Create a new folder and set up a virtual environment.

```
# Exercise 2: Set Up Your Environment
# Create a new folder and set up a virtual environment.
import os
folder_name = 'my_capstone_project'
os.makedirs(folder_name, exist_ok=True)
os.chdir(folder_name)
os.system('python -m venv my_project_env')
print('Folder created and virtual environment set up in:', folder_name)
```
